```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

combined_features = pd.read_csv('data/Shop_Features.csv')
```

```python
combined_features.head()
```

```python
combined_features.info()
```

```python
combined_features.describe()
```

```python
sns.histplot(data=combined_features, x='total_sales', kde=True)
plt.show()

sns.histplot(data=combined_features, x='total_quantity_sold', kde=True)
plt.show()

sns.histplot(data=combined_features, x='total_transactions', kde=True)
plt.show()
```

```python
# Select only numeric columns
numeric_columns = combined_features.select_dtypes(include=['number'])

# Calculate correlations
correlations = numeric_columns.corr()

# Plot the heatmap
sns.heatmap(correlations, annot=True, cmap='coolwarm')
plt.show()
```

```python
sns.scatterplot(data=combined_features, x='total_sales', y='total_transactions')
plt.show()

sns.scatterplot(data=combined_features, x='total_sales', y='shop_area_sq_ft')
plt.show()

sns.scatterplot(data=combined_features, x='total_quantity_sold', y='total_unique_items')
plt.show()
```

```python
sns.boxplot(data=combined_features, x='total_weeks', y='total_sales')
plt.show()

sns.boxplot(data=combined_features, x='total_weeks', y='total_quantity_sold')
plt.show()

sns.boxplot(data=combined_features, x='total_weeks', y='total_transactions')
plt.show()
```

```python
sns.pairplot(combined_features)
plt.show()
```
